```python
import clingo
```

```python


class Context:
    c = 0

    @staticmethod
    def log(message):
        Context.c += 1
        print("[LOG]:", Context.c, message)
        return clingo.Number(Context.c)
```

```python
instance = """

#program first.
p(@log(first)).

#program second.
q(@log(second)).

#program third.
u(@log(yes)) :- s(_).
t(@log(no)) :- s(0).
s(@log(third)) :- p(1), q(2).
"""
```

```python
Context.c = 0
ctl = clingo.Control(("--models", "0"))
ctl.add("base", [], instance)
ctl.ground((('first',[]), ('second', []), ('third', []), ('base', [])), context=Context)
ctl.solve(on_model=print)
```

```python
Context.c = 0
ctl = clingo.Control(("--models", "0"))
ctl.add("base", [], instance)
ctl.ground((('first', []),), context=Context)
ctl.ground((('second', []),), context=Context)
ctl.ground((('third', []),), context=Context)
ctl.ground((('base', []),), context=Context)
ctl.solve(on_model=print)
```
